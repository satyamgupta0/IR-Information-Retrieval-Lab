```python
import nltk
import scipy.stats as ss
import matplotlib.pyplot as plt 
import math
import re
from nltk.stem import PorterStemmer
from wordcloud import WordCloud, STOPWORDS
```

```python
from nltk.corpus import reuters
```

```python
nltk.download('wordnet')
nltk.download('reuters')
```

```python
wordnet_words = [w.lower() for w in reuters.words()]
words = set(wordnet_words)
# counts = [(w, wordnet_words.count(w)) for w in words]
count = {}
for word in wordnet_words:
  count[word] = count.get(word,0) + 1
counts = []
for k,v in count.items():
  counts.append((k,v))
```

```python
words_1 = [w for (w, c) in counts]
counts_1 = [c for (w, c) in counts]
# creating a dictionary to generate workd
dictIs = {}
for i in range(len(words_1)):
  dictIs[words_1[i]] = counts_1[i]; 
```

```python
count_rank = ss.rankdata([c for (w, c) in counts], method = 'ordinal')
count_rank
```

```python
counts_ranked = zip(words_1, counts_1, count_rank)
#counts_ranked[100:110]
```

```python
counts_sorted = sorted(counts_ranked, key=lambda item: item[2])
counts_sorted[20000:20010]
```

```python
counts_sorted[-50:]
```

```python
l = [word[1] for word in counts_sorted[-50:]]
plt.figure(figsize=(14,10))
plt.hist(l,bins=20)
plt.show()
```

```python
L = len(counts_sorted)
plt.plot([math.log(L-r+1) for (w, c, r) in counts_sorted], [math.log(c) for (w, c, r) in counts_sorted], 'ro')
```

```python
import numpy as np
```

```python
# generate x and y
x = [math.log(L-r+1) for (w, c, r) in counts_sorted]
y = [math.log(c) for (w,c,r) in counts_sorted]
X = np.array(x)
Y = np.array(y)
A = np.vstack([X, np.ones(len(X))]).T
Y_ = Y[:, np.newaxis]
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),Y_)
print(alpha)
```

```python
plt.plot(X, Y, 'b.')
plt.plot(X, alpha[0]*X + alpha[1], 'r')
plt.xlabel('log(rank)')
plt.ylabel('log(frequency)')
plt.show()
```

```python
# Buiding a word cloud
wc = WordCloud(background_color = "white", max_words = 1000);
wc.generate_from_frequencies(dictIs)
plt.imshow(wc, interpolation = 'bilinear');
plt.axis('off');
plt.show()
```

```python
nivas = PorterStemmer()
def remove_symbols(line):
  print(re.sub('[^A-Za-z0-9\s]+', '', line).lower())
  return re.sub('[^A-Za-z0-9\s]+', '', line).lower()
stopwords = set(STOPWORDS)
freq_dict = {}
for w, c in counts:
  if w not in stopwords:
    w = remove_symbols(w)
    if w:
      w = nivas.stem(w)
      freq_dict[w] = freq_dict.get(w, 0) + c
wc = WordCloud(background_color="white")
wc.generate_from_frequencies(freq_dict)
plt.imshow(wc, interpolation = 'bilinear');
plt.axis('off');
plt.show()
```
